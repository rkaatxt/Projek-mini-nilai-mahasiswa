## MEMBUAT DATASET 'data_nilai.csv'

```python
import csv
data_mahasiswa = [
    ['Nim', 'Nama', 'Alpro', 'Varkom', 'Ralis'],
    ['124400047', 'Hu Tao', 80, 75, 85],
    ['124400012', 'Phrolova', 60, 50, 60],
    ['124400023', 'Furina', 90, 95, 90],
    ['124400034', 'Bennett', 50, 40, 50],
    ['124400045', 'camellya', 85, 80, 88],
    ['124400056', 'Chisa', 70, 70, 75],
    ['124400067', 'Paimon', 45, 30, 40],
    ['124400078', 'Ruan Mei', 95, 90, 95]
]
with open('data_nilai.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data_mahasiswa)

print("File 'data_nilai.csv' berhasil dibuat untuk simulasi.")
```

## IMPORT LIBRARY

Library diimpor pada awal program untuk menyediakan fungsi-fungsi pendukung yang dibutuhkan. Pada projek ini, library `csv` digunakan untuk membaca dan menulis data nilai mahasiswa dalam format CSV sehingga proses pengolahan data dapat dilakukan secara efektif dan terstruktur.

```python
import csv
import matplotlib.pyplot as plt
```

## LOAD DATA CSV

Load Data CSV adalah proses membaca data nilai mahasiswa dari file CSV ke dalam program Python menggunakan library `csv`. Data dibaca dalam bentuk dictionary dan disimpan ke dalam list agar mudah diolah dan dianalisis.

```python
def load_data(nama_file):
    data = []
    with open(nama_file, newline='') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(row)
    return data

data_mahasiswa = load_data("data_nilai.csv")
data_mahasiswa
```

## PREPROCESSING

Preprocessing dilakukan untuk mengubah tipe data nilai mahasiswa dari string menjadi integer agar dapat digunakan dalam proses perhitungan dan analisis data.

```python
for mhs in data_mahasiswa:
    mhs["Alpro"] = int(mhs["Alpro"]) 
    mhs["Varkom"] = int(mhs["Varkom"])
    mhs["Ralis"] = int(mhs["Ralis"])
    
data_mahasiswa
```

## FUNGSI (DEF & LAMDA)

Fungsi def digunakan untuk perhitungan yang lebih kompleks seperti menghitung rata-rata nilai, sedangkan fungsi lambda digunakan untuk operasi sederhana seperti menentukan predikat berdasarkan nilai rata-rata.

```python
def hitung_rata_rata(mhs):
    return (mhs["Alpro"] + mhs["Varkom"] + mhs["Ralis"]) / 3

predikat = lambda rata: (
    "A" if rata >= 75 else
    "AB" if rata >= 70 else
    "B" if rata >= 65 else
    "BC" if rata >= 60 else
    "C" if rata >= 55 else
    "D"
)

contoh_mhs = {
    "nama": "Hu Tao",
    "Alpro": 80,
    "Varkom": 75,
    "Ralis": 85
}

rata = hitung_rata_rata(contoh_mhs)
print("Nama Mahasiswa :", contoh_mhs["nama"])
print("Rata-rata Nilai:", round(rata, 2))
print("Predikat       :", predikat(rata))
```

## OOP

OOP digunakan untuk merepresentasikan mahasiswa sebagai objek yang memiliki data dan fungsi, sehingga pengolahan nilai menjadi lebih terstruktur dan mudah dikembangkan.

```python
class Mahasiswa:
    def __init__(self, nim, nama, alpro, varkom, ralis):
        self.nim = nim
        self.nama = nama
        self.alpro = int(alpro)
        self.varkom = int(varkom)
        self.ralis = int(ralis)
        self.tambahan = 0 

    def hitung_rata_rata(self):
        return (self.alpro + self.varkom + self.ralis) / 3

    def get_predikat(self):
        rata = self.hitung_rata_rata()
        if rata >= 75: return "A"
        elif rata >= 70: return "AB"
        elif rata >= 65: return "B"
        elif rata >= 60: return "BC"
        elif rata >= 55: return "C"
        else: return "D"

    def get_status_kelulusan(self):
        return "LULUS" if self.hitung_rata_rata() >= 55 else "TIDAK LULUS"

mhs1 = Mahasiswa("124400047", "Hu Tao", 80, 75, 85)
print("Nama:", mhs1.nama)
print("Rata-rata:", mhs1.hitung_rata_rata())
print("Predikat:", mhs1.get_predikat())
print("Status:", mhs1.get_status_kelulusan())
```

## PERCABANGAN & PERULANGAN

Perulangan digunakan untuk memproses data seluruh mahasiswa, sedangkan percabangan digunakan untuk menentukan predikat dan status kelulusan berdasarkan nilai rata-rata.

```python
for mhs in data_mahasiswa:
    rata = hitung_rata_rata(mhs)
    mhs["rata_rata"] = rata
    mhs["predikat"] = predikat(rata)

    if rata >= 55:
        status = "LULUS"
    else:
        status = "TIDAK LULUS"

    print(mhs["Nama"], "| Rata-rata:", round(rata,2), "| Predikat:", mhs["predikat"], "|", status)
```

## EXCEPTION HANDLING

Exception Handling merupakan mekanisme dalam Python yang digunakan untuk menangani kesalahan (*error*) yang dapat terjadi saat program dijalankan tanpa menghentikan keseluruhan program. Dengan menggunakan exception handling, program tetap dapat berjalan meskipun terjadi kesalahan pada salah satu bagian proses. Pada projek analisis nilai mahasiswa, exception handling digunakan untuk mengantisipasi kesalahan yang mungkin terjadi, seperti kesalahan input dari pengguna atau kesalahan konversi tipe data. Kesalahan tersebut biasanya muncul ketika input yang dimasukkan tidak sesuai dengan tipe data yang diharapkan, misalnya apabila pengguna memasukkan teks saat program meminta input berupa angka. Penggunaan struktur `try` dan `except` memungkinkan program untuk menjalankan kode yang berpotensi menimbulkan error pada blok `try`. Apabila terjadi kesalahan, program tidak langsung berhenti, tetapi akan mengeksekusi blok `except` untuk menampilkan pesan kesalahan yang informatif kepada pengguna. Dengan demikian, exception handling membantu meningkatkan keandalan dan kenyamanan penggunaan program.

```python
for mhs in data_mahasiswa:
    try:
        tambahan = int(input(f"Masukkan nilai tambahan untuk {mhs['Nama']}: "))
        mhs["tambahan"] = tambahan
    except ValueError:
        print("Input harus berupa angka, nilai tambahan di-set 0")
        mhs["tambahan"] = 0
```

## VISUALISASI DATA

Visualisasi data digunakan untuk menyajikan hasil analisis nilai mahasiswa dalam bentuk grafik agar lebih mudah dipahami dan dianalisis.

```python
nama = [mhs["Nama"] for mhs in data_mahasiswa]
rata = [mhs["rata_rata"] for mhs in data_mahasiswa]

plt.figure()
plt.bar(nama, rata)
plt.title("Rata-rata Nilai Mahasiswa")
plt.xlabel("Nama Mahasiswa")
plt.ylabel("Nilai")
plt.show()
```
